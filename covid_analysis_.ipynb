```python
# Global totals per date
global_daily = df.groupby('date_number')[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()

# Plot global growth curves
plt.figure(figsize=(12, 6))
plt.plot(global_daily['date_number'], global_daily['Confirmed'], label='Confirmed')
plt.plot(global_daily['date_number'], global_daily['Deaths'], label='Deaths')
plt.plot(global_daily['date_number'], global_daily['Recovered'], label='Recovered')
plt.xlabel("Date")
plt.ylabel("Count")
plt.title("Global COVID-19 Trend Over Time")
plt.legend()
plt.show()
```
